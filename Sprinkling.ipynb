```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
n = 500
root2 = 1/np.sqrt(2)
points, cartesian = np.random.random_sample((n, 2)), np.random.random_sample((n, 2))
relations, links = np.zeros((n,n)), np.zeros((n,n))
for i, p1 in enumerate(points): 
    for j, p2 in enumerate(points): 
        relations[i,j] = 1 if p1[0] < p2[0] and p1[1] < p2[1] else 0
R2 = np.matmul(relations, relations)
for i, p1 in enumerate(points): 
    for j, p2 in enumerate(points): 
        links[i,j] = relations[i,j]-1 if R2[i,j] > 0 else relations[i,j]
for _, (u, v) in enumerate(points):
    cartesian[_] =  root2*(v-u), root2*(v+u)
```

```python
%matplotlib inline
fig, ax = plt.subplots(1,1,figsize=(8,8))
ct = np.transpose(cartesian)
plt.plot(ct[0], ct[1], 'o', mew=1, color='salmon')
plt.plot([0,root2,0,-root2,0], [0,root2,2*root2,root2,0], linestyle='-', linewidth=3, color='b')
plt.axis('off')
plt.show()
```

```python
n = 200
root2 = 1/np.sqrt(2)
xcoord = np.random.uniform(-root2,root2,n)
tcoord = [np.random.uniform(-root2+np.abs(x),root2-np.abs(x)) for x in xcoord]
cartesian = list(zip(xcoord,tcoord))
theta = [2*np.pi*_ for _ in np.random.random_sample((n))]
cartesian_3d = [[r*np.cos(theta[_]), r*np.sin(theta[_]), t] for _, (r,t) in enumerate(cartesian)]
```
